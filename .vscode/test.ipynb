```python
list1 = ['a','b','c','d']

list1.append('e')
print(list1)

list1.extend(['ABC','DEC'])
print(list1)

list1.insert(1,"XYZ")
list1.pop()
print(list1)

list1.pop(1)
print(list1)

list1.remove('ABC')
print(list1)

h = set(list1)
print(h)
```

```python
list3 = [1,1,2,2,3,3]
print(list3)

s = set(list3)
print(s)

h = tuple(list3)
print(h)

s2 = set('abbcccddss')
# A set stores only unique values (no duplicates)
print(s2)
```
